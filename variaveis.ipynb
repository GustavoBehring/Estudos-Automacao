```python
faturamento = 1000
custo = 500
lucro = faturamento - custo

print(f"O faturamento da loja é de {faturamento} e o lucro de {faturamento - custo}")  #f-string
print(f"O faturamento da loja é de {faturamento:.2f}")  #f-string
print("O faturamento da loja foi de " + str(faturamento))
print("O faturamento da loja foi de {}. O custo da loja foi de {}. Assim, o lucro da loja foi {}".format(faturamento, custo, lucro))
```
