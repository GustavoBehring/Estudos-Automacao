# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e 
do teclado

```python
#necessario instalação das bibliotecas
#pyautogui -> automações
#pandas -> manipulação de dados
# openpyxl -> trabalhar com bases de dados excel
# para instalar: abrir o terminal e digitar: pip install nome_da_biblioteca


#pyautogui.click -> cliclar com o mouse
#pyautogui.write -> digitar com o teclado
#pyautogui.press -> apertar uma tecla do teclado
#pyautogui.hotkey -> apertar uma combinação de teclas
```

```python
import pyautogui
import time




pyautogui.PAUSE = 1  # tempo de espera entre as ações

# Passo a passo do desafio 
# Passo 1: Entrar no sistema da empresa (link do drive)
# abrir o Opera
pyautogui.press("win")
pyautogui.write("Opera")
pyautogui.press("enter")


time.sleep(2)  # esperar o navegador abrir

# Digite o link do sistema
link = "https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga"
# a esquerda do igual é a variavel e a direita o valor
pyautogui.write(link)

# apertei enter e esperei
pyautogui.press("enter")
time.sleep(2)





```

```python
# Passo 2: Navegar no sistema para encontrar a base de dados
pyautogui.click(x=392, y=359, clicks=2)
time.sleep(2)


# Passo 3: Exporar a base de dados (baixar o arquivo)
pyautogui.click(x=2456, y=360) 
pyautogui.click(x=2239, y=443)

time.sleep(5)

```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
import pandas

# Passo 4: Calcular os indicadores (faturamento e quantidade de produtos vendidos)
# abrir a base de dados
caminho = r"C:\Users\gutyb\Downloads\Vendas - Dez.xlsx"  #r -> raw string para evitar erros com barras
tabela = pandas.read_excel(caminho)

# ver as informações da base de dados
display(tabela)
# somar o faturamento de todos os produtos = somar valor final
faturamento = tabela["Valor Final"].sum()
# somar a quantidade de produtos = somar a quantidade vendida
qtde_produtos = tabela["Quantidade"].sum()

print(faturamento)
print(qtde_produtos)





```

### Vamos agora enviar um e-mail pelo gmail

```python
import pyperclip

# Passo 5: Enviar as informações por e-mail 
# abrir nova aba

pyautogui.hotkey("ctrl", "t")  

# entrar no gmail

pyautogui.write("outlook.com") 
pyautogui.press("enter")      
time.sleep(3)  # esperar abrir o email

# clicar no botão escrever novo email
pyautogui.click(x=171, y=221)  
time.sleep(1)  # esperar abrir a tela de novo email

#para quem enviar o email  
pyperclip.copy("guty.behring+código@hotmail.com")
pyautogui.hotkey("ctrl", "v")
#pyautogui.press("tab")  #seleciona o e-mail
pyautogui.press("tab")  #vai para o campo do assunto



#qual o assunto
pyperclip.copy("Relatório de Vendas")   #usar o pyperclip para copiar o texto para a área de transferência pois não consegue escrever carateres especiais
pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab")  #vai para o campo do corpo do email

#corpo do email
texto = f"""
Prezados, bom dia

Segue o relatório das vendas de hoje.

Faturamento: R${faturamento:,.2f} 
Quantidade de produtos vendidos: {qtde_produtos:,}

Qualquer dúvida estou à disposição.
Atenciosamente,
Gustavo Behring
"""
pyperclip.copy(texto)           #f = float e é casa decimal e , separador de milhar
pyautogui.hotkey("ctrl", "v")

#enviar o email
pyautogui.click(x=803, y=282)
```
